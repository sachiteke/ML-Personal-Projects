1. Load and analyze data

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import mglearn

data = load_iris()
x_train, x_test, y_train, y_test = train_test_split(data['data'],data['target'],random_state=1)

df = pd.DataFrame(x_train,columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
matrix = pd.plotting.scatter_matrix(df, c=y_train, figsize=(15,15), marker='o', hist_kwds= {'bins':60}, s=60, alpha=1, cmap=mglearn.cm3)
```

2. Build the model

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

3. Evaluate the model

```python
output = knn.predict(x_test)
print("Test score: ", np.mean(output == y_test))
print("Test set score: ", knn.score(x_test, y_test))
```

4. Test the model

```python
# For versicolor

test_set = np.array([[6.3, 2.9, 4.2, 0.3]])
prediction = knn.predict(test_set)
print(data['target_names'][prediction])
```
