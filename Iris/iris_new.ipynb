1. Load the data and analyze it

```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
import mglearn
import numpy as np

data = pd.read_csv("Iris.csv", header=None)
data = data.values
x = data[1:,1:5]
y = []
for i in range(3):
    for j in range(50):
        y.append(i)

y = np.array(y)
x = x.astype('float64')
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1)

df = pd.DataFrame(x_train,columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
matrix = pd.plotting.scatter_matrix(df, c=y_train, figsize=(15,15), marker= 'o', hist_kwds= {'bins':20}, s=60, alpha=1)
```

2. Build the model

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

3. Evaluate the model

```python
output = knn.predict(x_test)
print(output)
print("Test score: ", np.mean(output==y_test))
print("Test score: ", knn.score(x_test,y_test))
```

4. Test the model

```python
# For iris-setosa

testing = np.array([[2.5,1.5,1.0,0.4]])
prediction = knn.predict(testing)
print(data[1:,5:][prediction])
```
